```python
import pandas as pd
```

```python
patients = pd.read_csv("patients.csv")
conditions = pd.read_csv("conditions.csv")
medications = pd.read_csv("medications.csv")
immunizations = pd.read_csv("immunizations.csv")
procedures = pd.read_csv("procedures.csv")
```

```python
paset = set(patients.columns.to_list())
```

```python
cset = set(conditions.columns.to_list())
```

```python
mset = set(medications.columns.to_list())
```

```python
iset = set(immunizations.columns.to_list())
```

```python
prset = set(procedures.columns.to_list())
```

```python
all_set = paset | cset | mset | iset | prset
```

in this project, i developed a ELT pipeline that loads various csv files (patients.csv, immunizations.csv, conditions.csv, procedures.csv) to a "raw" schema in a dockerized postgres database that has a named volume to persist the data. i then used dbt to test the quality of the data and clean the data for reporting. the reporting is via a streamlit dashboard. in dlt, i enforce schema using pydantic.
